# Standard Data Cleansing

```python
# Finding missing value null #
df.isna().sum()
```

```python
# Check spasi in field #
df[df['TotalCharges'].str.contains(pat=' ', case=False)]
```

```python
# Finding duplicate data #
df.duplicated().sum()
```

## Univariate Analysis

#### Boxplot to detect outliers

```python
# adjust the figure size for better readability
plt.figure(figsize=(12,6))

# plotting
features = numericals
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=df[features[i]], color='cyan')
    plt.tight_layout()
```

### KDE plot for knowing the distribution form

```python
# adjust the figure size for better readability
plt.figure(figsize=(15,6))

features = numericals
for i in range(0, len(features)):
    plt.subplot(2, len(features)//2 + 1, i+1)
    #plt.subplot(1, len(features), i+1)
    sns.distplot(x=df[features[i]], color='skyblue')
    plt.xlabel(features[i])
    plt.tight_layout()
```

Observation:
1. We can ignore interpreting feature columns with limited discrete values such as SeniorCitizen
2. The true continuous numeric features: tenure, MonthlyCharges, and TotalCharges haven't normal distribution (symmetric). 
*   tenure and monthlycharges are bimodal distribution

*  TotalCharges is positive skew
Therefore, need to approximate to normal distribution using smote
3. Most customers are (distribution peak)


*   Not Senior Citizen (based on the peak 0 on the graph)
*   Monthly Charges and Total Charges are have the same peak > $18
*   tenure >= 1

## 1. Top 10 customers that have the most total charges/ total spend

```python
# group by aggregation of top 10 customer #
top10_cust = (df
            .groupby('customerID')
            .agg(Total_Charges=('TotalCharges','sum'))
            .reset_index()
            .sort_values('Total_Charges',ascending=False)
            .head(10)
            )

top10_cust
```

```python
# visualize top 10 customer #
plt.figure(figsize=(10,10))
top10_cust['customerID'] = top10_cust['customerID'].astype(str)
sns.barplot(data=top10_cust, x='customerID', y='Total_Charges')
plt.xticks(rotation=90)
```

Observation 
1. From the graph, we know that top 10 customer have approximate same amount in total charges.
2. From this top 10 customer, the data will observe more

```python
# lookup the data of top 10 customer to observe #
df[df['TotalCharges'] >= 8477.70]
```

observation :

## 2. Does contract affect total charges ?

```python
# group by aggregation of contract and total charges #
Top3_Contract = (df
            .groupby('Contract')
            .agg(Total_Charges=('TotalCharges','sum'))
            .reset_index()
            .sort_values('Total_Charges',ascending=False)
            )

Top3_Contract
```

```python
# visualize top 3 contract #
plt.figure(figsize=(10,10))
Top3_Contract['Contract'] = Top3_Contract['Contract'].astype(str)
sns.barplot(data=Top3_Contract, x='Contract', y='Total_Charges')
plt.ylim(0,7000000) 
```

Observation :
1. From the graph, all of the contract with 2 years have the highest total charges than the others.
2. Next, will figure out teh plot of total charges per customer ID

```python
# Get the data of contract, customer id and total charges #
Contract_charges = (df
                .groupby(['Contract','customerID'])
                .agg(Total_Charges=('TotalCharges','sum'))
                .reset_index())
Contract_charges.head()
```

```python
# Look the plot of contract and total charges of the customers #
sns.boxplot(data=Contract_charges,
            x='Contract',
            y='Total_Charges')
```

Observation :
1. the highest median on total charges is two years contract than the others.
2. There's outliers of total charges on contract month-to-month that indicate customers have to retention the subscription every-month.
